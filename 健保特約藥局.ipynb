```python
import pandas as pd

url = "https://data.nhi.gov.tw/Datasets/DatasetResource.ashx?rId=A21030000I-D21005-004"
df = pd.read_csv(url, encoding="utf8")
df.head()
```

```python
df3 = df[df["終止合約或歇業日期 "].isna()]
df3["地 址 "].isna().sum()
```

```python
df2 = df[df["終止合約或歇業日期 "].isna()]
df2.dropna(subset=["地 址 "], inplace=True)
```

```python
addr = input("請輸入部分地址:")

df2[df2["地 址 "].str.contains(addr)]
```

```python
df2.loc[4:4]
```

```python
import geocoder # pip install geocoder
for addr in df2.loc[4:4, "地 址 "]:
    print(addr)
    g = geocoder.arcgis(addr)
    print(g.json)
    print(g.json["address"].split(",")[1].strip())
```

```python
for addr in df.loc[4:4, "地 址 "]:
    g = geocoder.arcgis(addr)
    print(g.json["lat"])
    print(g.json['lng'])
```
